# Modelo de aprendizado de máquina para Íris

Importar bibliotecas


```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier 
from sklearn.metrics import accuracy_score, classification_report
import numpy as np
```

Dataset Iris


```python
# 1. Baixar o conjunto de dados no GitHub
# 2. Fazer a leitura dos dados utilizando Pandas

# Arquivo CSV
df = pd.read_csv("https://raw.githubusercontent.com/marcelovca90-inatel/AG2/refs/heads/main/iris.csv") 
    
print("DataFrame Iris:")
print(df.head())
```

Mapeamento

```python
# 3. Mapeamento para conversão:
# Iris-setosa=1, Iris-versicolor=2, Iris-virginica=3
mapeamento_especies = {
    'Iris-setosa': 1,
    'Iris-versicolor': 2,
    'Iris-virginica': 3
}

# Converter os valores da coluna 'species' para inteiro
df['species'] = df['species'].replace(mapeamento_especies)

print("\nDataFrame após a conversão de 'species' (últimas 5 linhas):")
print(df.tail())
print("\nTipo de dado para 'species' após a conversão:", df['species'].dtype)
```

Separar features (X) e (Y) classes

```python
# Separar features (X) e target (y)
X = df.drop('species', axis=1) # Colunas das 4 medidas
y = df['species']             # Coluna espécies

# 4. Separar e embaralhar os dados (80% treino, 20% teste) 
# O argumento shuffle=True é o padrão. random_state garante a reprodutibilidade.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True, random_state=42)

print(f"Tamanho do conjunto de Treinamento: {len(X_train)} amostras (80%)")
print(f"Tamanho do conjunto de Teste: {len(X_test)} amostras (20%)")
```

Treinar usando o modelo KNN

```python
# Usando k=5 (valor inicial escolhido)
k_value = 5
modelo_knn = KNeighborsClassifier(n_neighbors=k_value) 

# 5. Treinar o modelo escolhido usando 80% dos dados.
modelo_knn.fit(X_train, y_train) 

print(f"\nModelo k-NN (k={k_value}) treinado com sucesso!")
```

Avaliar o modelo

```python
# 6. Avaliar o modelo escolhido usando os 20% restantes.
y_pred = modelo_knn.predict(X_test)

# 7. Exibir métricas de avaliação.
acuracia = accuracy_score(y_test, y_pred)
relatorio = classification_report(y_test, y_pred)

print("\n--- Métricas de Avaliação do Modelo k-NN ---")
print(f"Acurácia: {acuracia*100:.2f}%")
print("\nRelatório de Classificação:\n", relatorio)
```

Predição por entrada
